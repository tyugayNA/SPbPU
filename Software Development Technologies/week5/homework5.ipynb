```python
import pandas as pd
```

```python
df_covidinfo = pd.DataFrame(info)
df_covidinfo
```

```python
import plotly.graph_objs as go
from plotly.offline import iplot  #iplot only works with this code (use offline)
```

```python
normdate = pd.to_datetime(df_covidinfo.date)
normdate
```

```python
trace1 = go.Scatter( x = list(range(len(info))),
                    y = df_covidinfo.positive,
                    mode = "lines",
                    name = "covid positive",
                    marker = dict(color = "Red"),
                    text = df_covidinfo.positive)
trace2 = go.Scatter( x = list(range(len(info))),
                    y = df_covidinfo.negative,
                    mode = "lines",
                    name = "covid negative",
                    marker = dict(color = "Blue"),
                    text = df_covidinfo.negative)
trace3 = go.Scatter( x = list(range(len(info))),
                    y = df_covidinfo.hospitalizedCumulative,
                    mode = "lines",
                    name = "covid hospitalized",
                    marker = dict(color = "Yellow"),
                    text = df_covidinfo.hospitalizedCumulative)
dataMy = [trace1, trace2, trace3]
layoutMy = dict(title = 'Covid information',
              xaxis= dict(title= 'Days')
             )
fig = dict(data = dataMy, layout = layoutMy)
iplot(fig)
```
