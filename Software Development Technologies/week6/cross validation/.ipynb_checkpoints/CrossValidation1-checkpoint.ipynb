```python
# create the range 1 to 25
rn = range(1,26)
```

```python
from sklearn.model_selection import KFold
#kf5 = KFold(n_splits=10, shuffle=False)
kf3 = KFold(n_splits=5, shuffle=True,random_state=500) #, shuffle=True,random_state=4210
# create two Kfolds, one splitting data 3-times and other doing 5 folds.
```

```python
# the Kfold function retunrs the indices of the data. Our range goes from 1-25 so the index is 0-24
for train_index, test_index in kf3.split(rn):
    print(train_index, test_index)
```

```python
# to get the values from our data, we use np.take() to access a value at particular index
import numpy as np
for train_index, test_index in kf3.split(rn):
    print(np.take(rn,train_index), np.take(rn,test_index))
```

```python
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
#DONT eat sugar its bad for ur health
```

```python
# Load the Diabetes dataset
columns = "age sex bmi map tc ldl hdl tch ltg glu".split() # Declare the columns names
diabetes = datasets.load_diabetes() # Call the diabetes dataset from sklearn
df = pd.DataFrame(diabetes.data, columns=columns) # load the dataset as a pandas data frame
y = diabetes.target # define the target variable (dependent variable) as y
```

```python
df.head()
```

```python
# create training and testing vars
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.3)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# fit a model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```
