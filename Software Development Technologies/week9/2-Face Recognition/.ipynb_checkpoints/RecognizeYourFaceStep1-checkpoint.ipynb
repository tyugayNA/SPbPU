```python
import zipfile
path = 'OG.zip'  
zip_object = zipfile.ZipFile(file=path, mode='r')
zip_object.extractall('./')
zip_object.close()
```

```python
def get_image_data():
  paths = [os.path.join('C:/Users/MONSTER/MACHINE LEARNING/2022_New_Researches/ADVANCED COMPUTER VISION/2-Face Recognition/OG', f) for f in os.listdir('C:/Users/MONSTER/MACHINE LEARNING/2022_New_Researches/ADVANCED COMPUTER VISION/2-Face Recognition/OG')] #SAME PATH
  faces = []
  ids = []
  for path in paths:
    image = Image.open(path).convert('L')
    image_np = np.array(image, 'uint8')
    id = int(path.split('.')[1])
    
    ids.append(id)
    faces.append(image_np)
  
  return np.array(ids), faces
```

```python
import os
ids, faces = get_image_data()
```

```python
lbph_classifier = cv2.face.LBPHFaceRecognizer_create() # make gray scale selfies  (approximately 30)
lbph_classifier.train(faces, ids)
lbph_classifier.write('C:/Users/MONSTER/MACHINE LEARNING/2022_New_Researches/ADVANCED COMPUTER VISION/2-Face Recognition/lbph_classifierOG3.yml')  #HERE I WRITE A CLASSIFIER, YOU CAN SEND YOUR PICTURES AND MAKE YOUR OWN CLASSIFIER
```

```python
paths = [os.path.join('C:/Users/MONSTER/MACHINE LEARNING/2022_New_Researches/ADVANCED COMPUTER VISION/2-Face Recognition/OG/', f) for f in os.listdir('C:/Users/MONSTER/MACHINE LEARNING/2022_New_Researches/ADVANCED COMPUTER VISION/2-Face Recognition//OG/')]
for path in paths:
  image = Image.open(path).convert('L')
  image_np = np.array(image, 'uint8')
  prediction, _ = lbph_face_classifier.predict(image_np)
  expected_output = int(path.split('.')[1])

  cv2.putText(image_np, 'Pred: ' + str(prediction), (10,30), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0,255,0))
  cv2.putText(image_np, 'Exp: ' + str(expected_output), (10,50), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0,255,0))
```
