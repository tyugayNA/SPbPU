We'll start by loading up the MovieLens dataset. Using Pandas, we can very quickly load the rows 
#have you seen it,its a cool movie starring Al Pacino and my favorite Action movies actor Keanu Reeves

```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating']  #  keeps ratings here
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']  # i keep title here
movies = pd.read_csv('ml-100k/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings) ## merging both of them below
```

```python
ratings.head()
```

Now the amazing pivot_table function on a DataFrame will construct a user / movie rating matrix. Note how NaN indicates missing data - movies that specific users didn't rate.

```python
movieRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movieRatings.head()  # pivot table creates a table how i want
# obviously, not all the users watched the all films, thats why lots of NANS 
# you asked for this prepare yourself FOR TORTURE
# so now, we can extract vectors easily( Colabrative filtering- vectors good)
pd.set_option('display.max_columns', None)
movieRatings.head()
```

erxtract a Series of users who rated my favorite film that inspired me to learn too many languages :P

```python
DevilRatings = movieRatings['Toy Story (1995)']  # exttract all the users who rated John Milton :D
DevilRatings.head()
```

Pandas' corrwith function makes it really easy to compute the pairwise correlation of films' vector of user rating with every other movie! After that, we'll drop any results that have no data, and construct a new DataFrame of movies and their correlation score (similarity) to Devil\'s Advocate:

```python
similarMovies = movieRatings.corrwith(DevilRatings)# checking correlation for each movie with  Devil's Advocate
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head(10)
```

Our results are probably getting messed up by movies that have only been viewed by a FEW of people. So we need to got rid of movies that were only watched by a few people that are producing bad results. 

```python
# the results were weird, so need to improve Recommending, heres how
import numpy as np
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]}) # aggregate everything with title 
movieStats.head()# rating size is important so i can get rid of smallers values,(affecting my correlation)
```

get rid of any movies rated by fewer than 100 people, and check the top-rated ones that are left:

```python
popularMovies = movieStats['rating']['size'] >= 100
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15] 
#i only take titles that has bigger than 100
```

 join this data with our original set of similar movies to  Devil's Advocate:

```python
df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))
# joining both ddatta( i can see similarity also) , used join operation
```
