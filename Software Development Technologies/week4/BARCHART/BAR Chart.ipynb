```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization   pip install seaborn  ---in anaconda prompt!!
sns.set() # this command sets the seaborn chart style as the default

```

```python
# Read the car data csv file for this task.
# Make sure the file is in the same directory, as your Python Notebook, otherwise you'll need to specify the file location.
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars.head(100)
```

```python
df_used_cars["Cars Listings"]
```

```python
plt.figure(figsize = (10, 5)) #determine the size of the chart
# To create a bar chart with matplotlib you can use the 'bar' function.
plt.bar(x = df_used_cars["Brand"], #specify the x axis
        height = df_used_cars["Cars Listings"], #specify the y axis
        color = "red") # the color for the bars

plt.xticks(rotation = 45,fontsize = 20) # rotate and format the labels for the x-axis
#rotation can be useful for you to fit the names of objects
plt.yticks(rotation = 45, fontsize = 20) # format the y-axis
plt.title("Cars Listings by Vladimr", fontsize = 15, fontweight = "bold") #add and format the title for the chart
plt.ylabel("Number of Listings", fontsize = 15 ) #add a title for the y-axis
plt.savefig("Used Cars Bar.png") # you can export your chart as a picture
plt.show() # depending on which environment you're using without this line you're chart may not display properly.
```

```python
mlist=["black","white","pink","purple","red","green","blue"]

plt.figure(figsize = (9, 6)) #determine the size of the chart
# To create a bar chart with matplotlib you can use the 'bar' function.
plt.bar(x = df_used_cars["Brand"], #specify the x axis
        height = df_used_cars["Cars Listings"], #specify the y axis
        color = mlist) 
plt.xticks(rotation = 45,fontsize = 13) # rotate and format the labels for the x-axis
#rotation can be useful for you to fit the names of objects
plt.yticks(rotation = 60,fontsize =13) # format the y-axis
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold") #add and format the title for the chart
plt.ylabel("Number of Listings", fontsize = 20 ) #add a title for the y-axis
plt.savefig("Used Cars Bar.jpg") # you can export your chart as a picture
plt.show() # depending on which environment you're using without this line you're chart may not display properly.
```
