```python
#first rule,, PIE charts are not so good, dont use them. But still this will help you to have it just in case!
#human brain is not that great understanding non-rectangular shapes
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default

#converts values into percentage format
```

```python
# load data. Number of cars based on their engine fuel types. Four engine fuel types in total : Diesel, Gas, Petrol and Other.
ogul = pd.read_csv("pie_chart_data.csv")
```

```python
ogul
```

```python
ogul['Engine Fuel Type'].values
```

```python
# create a pie chart with the pie function from matplotlib's pyplot.
# Labels are added as the number of cars in each category in the form of percentages.
# Additional labels formatting using textprops. Note it is a dictionary.  

plt.figure(figsize = (20, 7))
plt.pie(ogul['Number of Cars'], 
        labels = ogul['Engine Fuel Type'].values, 
        autopct = '%.2f%%',)
plt.legend()
plt.title('Vladimirs graph', fontsize = 50, fontweight = 'bold')
plt.show()
#you can change your text sizes, rotate them 
#check below
```

```python
# create a pie chart with the pie function from matplotlib's pyplot.
# Labels are added as the number of cars in each category in the form of percentages.
# Additional labels formatting using textprops. Note it is a dictionary.  

plt.figure(figsize = (10, 8))
plt.pie(ogul['Number of Cars'], 
        labels = ogul['Engine Fuel Type'].values, 
        autopct = '%.2f%%',  # float 
        textprops = {'size' : 'x-large',
                     'fontweight' : 'bold', 
                     'rotation' : '45',
                     'color' : 'k'}) #font color
plt.legend()
plt.title('Cars by Engine Fuel Type', fontsize = 18, fontweight = 'bold')
plt.show()
```
