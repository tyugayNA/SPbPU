# BARCHART

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv("BARCHART/bar_chart_data.csv")
df_used_cars
```

```python
plt.barh(df_used_cars["Brand"], df_used_cars["Cars Listings"],
         color=["#5a8278", "skyblue"], 
         edgecolor="Purple",
         linewidth=0.5
        )
plt.title("Cars", fontsize = 20, fontweight = "bold")
plt.xlabel("Cars Listings", fontsize = 15)
```

# HISTOGRAM 

```python
survey_data = pd.read_csv("HISTOGRAM/histogram_survey_data.csv")
survey_data
```

```python
#sns.set_style("dark")
plt.figure(figsize = (10,5))
plt.hist(survey_data["Age"],
         bins = survey_data["Age"].max() - survey_data["Age"].min(),
         color="#6d8a82")
plt.title("Geodemographic features", fontsize=20, fontweight="bold",color="#6d8a82")
plt.xlabel("Ages of people", fontsize=14)

fig, ax  = plt.subplots()
survey_data["Age"].plot.kde(ax=ax, title = "Geodemographic features")
survey_data["Age"].plot.hist(density=True, ax=ax)
plt.xlabel("Ages of people", fontsize=14)
ax.set_ylabel('Probability')

plt.show()
```

# STACKED AREA CHART

```python
area_chart = pd.read_csv("STACKED AREA CHART/area_chart_homework_data.csv")
area_chart.head(100)
```

```python
colors = ["#FF5733", "#8cb5bd", "#b38cbd"]
labels = ["International","Domestic","Services"]
plt.figure(figsize = (18, 6))
plt.title("Revenue for the last 12 months", fontsize = 14, weight = "bold", color="#5d7067")
plt.stackplot(area_chart["Date"],
              area_chart["International"],
              area_chart["Domestic"],
              area_chart["Services"],
             colors=colors)
plt.xticks(area_chart["Date"],rotation = 90)
plt.legend(labels = labels, loc = "upper left")
plt.show()
```
