```python
import pandas as pd

feature_names =  ['party','handicapped-infants', 'water-project-cost-sharing', 
                    'adoption-of-the-budget-resolution', 'physician-fee-freeze',
                    'el-salvador-aid', 'religious-groups-in-schools',
                    'anti-satellite-test-ban', 'aid-to-nicaraguan-contras',
                    'mx-missle', 'immigration', 'synfuels-corporation-cutback',
                    'education-spending', 'superfund-right-to-sue', 'crime',
                    'duty-free-exports', 'export-administration-act-south-africa']

voting_data = pd.read_csv('house-votes-84.data.txt', na_values=['?'], 
                          names = feature_names)
voting_data.head()
```

We can use describe() to get a summary about the data

```python
voting_data.describe()
```

drop missing values below(not always optimal to do that!( we dont have many missing values thus, it is OK to drop )

```python
voting_data.dropna(inplace=True)
voting_data.describe()
```

categorical values(strings, characters will not work with Neural networks, you need to convert them to meaningfull numerical values, here y and n => 1 and 0

```python
voting_data.replace(('y', 'n'), (1, 0), inplace=True)
voting_data.replace(('democrat', 'republican'), (1, 0), inplace=True)
```

```python
voting_data.head()
```

creating features and target(voting)

```python
all_features = voting_data[feature_names].drop('party', axis=1).values
all_classes = voting_data['party'].values
```

```python
#  use scikit_learn's cross_val_score to evaluate this model 
# Keras model in an estimator compatible with scikit_learn
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
from sklearn.model_selection import cross_val_score
def create_model():
    model = Sequential()
     
    model.add(Dense(16, input_dim=16, activation='relu'))
    model.add(Dense(32, activation='relu'))  
    model.add(Dense(16, activation='relu')) 
    model.add(Dense(16,activation='relu'))
    
    model.add(Dense(1,kernel_initializer='normal', activation='sigmoid')) # output layer    
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
estimator = KerasClassifier(build_fn=create_model, epochs=200, verbose=0)
cv_scores = cross_val_score(estimator, all_features, all_classes, cv=10)
cv_scores.mean()
```
